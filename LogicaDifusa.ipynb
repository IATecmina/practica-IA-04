```python
# practica-IA-04              By:: AY_SP
# Funcion de membresia para el subconjunto a
import matplotlib.pyplot as plt
import numpy as np

x = [0,1,8.2,8.2]
y = [0,1,1,0]

x2 = [0,8.3,9.8,9.8]
y2 = [0,1,1,0]

x3 = [0,9.9,16.4,16.4]
y3 = [0,1,1,0]

x4 = [0,16.5,32.8,32.8]
y4 = [0,1,1,0]

plt.plot(x,y)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.plot(x4,y4)
plt.title("Funcion de membresia - 1ra marcha")
plt.xlabel("% de pertenencia")
plt.ylabel("valores de velocidad")
plt.show()
```
